```python
import matplotlib.pyplot as plt
# import matplotlib.image as mpimg
import cv2
import numpy as np
image = cv2.imread('kodim05.png')
plt.imshow(image)
plt.show()
print("image: ", image.shape)

image_pad = np.zeros((512,1024,3)).astype(np.int) # .astype(uint8)
h, w = image.shape[:2]
image_pad[:h, :w] = image
image_pad = image_pad

plt.imshow(image_pad)
plt.show()
print("image: ", image_pad.shape)
# print(image_pad)

# import scipy.misc
# scipy.misc.imsave('outfile.jpg', image_pad)
```
